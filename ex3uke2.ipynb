```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import linregress
import matplotlib.pyplot as plt
from datetime import datetime
```

Forberedelse data

```python
df = pd.read_csv('svalbard_met.csv', sep=';', decimal=',')
df
```

Sletter alle rader med NaN verdier ved df.dropna()

```python
df = df.dropna()
df
```

```python
df = df.drop(range(16))
```

```python
df.head(20)
```

```python
y = df['Homogenisert middeltemperatur (mnd)']
y
```

```python
x = df['Tid(norsk normaltid)']
x
```

```python
x['Tid'] = pd.to_datetime(x['Tid'])
x['Tid_num'] = (x['Tid'] - x['Tid'].min()).dt.days
```

oppgave a - Use linear regression to determine the linear trend in the data from 1898 to 2023

```python
# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x,y)

# Create the regression line
regression_line = slope * y['Tid(norsk normaltid)_num'] + intercept

# Plot the data points and regression line
plt.scatter(y['Tid(norsk normaltid)'], y['Homogenisert middeltemperatur (mnd)'], label='Data')
plt.plot(y['Tid(norsk normaltid)'], regression_line, color='red', label='Linear Regression')

plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Linear Regression of Temperature over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
